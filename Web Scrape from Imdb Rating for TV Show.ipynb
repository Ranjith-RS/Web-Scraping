```python
import requests
from bs4 import BeautifulSoup
import openpyxl
```

```python
Headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36','Accept-Language':'en-US,en;-q=0.5'}
```

```python
url='https://www.imdb.com/chart/toptv/'
source=requests.get(url,headers=Headers)
```

```python
excel=openpyxl.Workbook()
sheet=excel.active
sheet.title="ImdbRating"
print(excel.sheetnames)
```

```python
sheet.append(['Show_rank','Show_Name','Released_year','Rating','Rated_by'])
```

```python
source.raise_for_status()
```

```python
soup=BeautifulSoup(source.text,'lxml')
```

```python
tag=soup.find('ul',class_='ipc-metadata-list ipc-metadata-list--dividers-between sc-3f13560f-0 sTTRj compact-list-view ipc-metadata-list--base')
print(len(tag))
```

```python
for i in tag:
    rank=i.find('h3',class_="ipc-title__text").text.strip(' ')[0]
    name=i.find('h3',class_="ipc-title__text").text.strip('.')[2:].lstrip()
    year=i.find('span',class_="sc-c7e5f54-8 hgjcbi cli-title-metadata-item").text
    rating=i.find('span',class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating").text.strip()[0:3]
    rated_by=i.find('span',class_="ipc-rating-star--voteCount").text.strip("()")[2:]
    print(rank,name,year,rating,rated_by)
    sheet.append([rank,name,year,rating,rated_by])
```

```python
excel.save('ImdbRating.xlsx')
```
